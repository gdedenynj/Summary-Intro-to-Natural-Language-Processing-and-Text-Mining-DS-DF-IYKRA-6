```python
df=pd.read_csv('dataset_tweet_sentimen_tayangan_tv.csv')
```

```python
df.head(2)
```

```python
#text preprocessing lowercase
df['Text Tweet']=[x.lower() for x in df['Text Tweet']]
```

```python
df['Text Tweet'][0]
```

```python
#remove whitespace
df['Text Tweet']=[x.strip() for x in df['Text Tweet']]
df['Text Tweet'][0]
```

```python
#tokenization
df['Text Tweet']=[x.split() for x in df['Text Tweet']]
df['Text Tweet'][0]
```

```python
#remove punctuation
remove_table=str.maketrans("","",punctuation)
nonpunctuation_stc=[]

for stc in df['Text Tweet']:
  words=[x.translate(remove_table) for x in stc]

  nonpunctuation_stc.append(words)

df['Text Tweet']=nonpunctuation_stc
df['Text Tweet'][0]
```

```python
#remove punctuation
remove_table=str.maketrans("","",punctuation)
nonpunctuation_stc=[]

for stc in df['Text Tweet']:
  words=[x.translate(remove_table) for x in stc]
  words2=[]

  for w in words:
    w=[x for x in w if x.isalpha()]
    w="".join(w)
    words2.append(w)

  nonpunctuation_stc.append(words2)

df['Text Tweet']=nonpunctuation_stc
df['Text Tweet'][0]
```

```python
#remove stopwords
nonstopwords=[]
for stc in df['Text Tweet']:
  words=[x for x in stc if x not in stop_words]

  nonstopwords.append(words)
```

```python
df['Text Tweet'] = nonstopwords
df['Text Tweet'] = [" ".join(x) for x in df['Text Tweet']]
df['Text Tweet']
```
